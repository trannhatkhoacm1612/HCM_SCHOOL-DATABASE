# Các hàm ghi file


```python
def file_loaitruong(dataframe,output_file): # import loaitruong
    with open(output_file,"w",encoding="utf-8") as file:
        file.write("use truonghoc;\n")
        for index,row in dataframe.iterrows():
            [maloaitrng,tenloaitrng] = list(row)
            querry = f'INSERT INTO loaitruong(maloaitrng,tenloaitrng) values("{maloaitrng}","{tenloaitrng}");\n'
            file.write(querry)
```

```python
def file_loaihinh(dataframe,output_file): # import loaihinh
    with open(output_file,"w",encoding="utf-8") as file:
        file.write("use truonghoc;\n")
        for index,row in dataframe.iterrows():
            [maloaihnh,tenloaihnh] = list(row)
            querry = f'INSERT INTO loaihinh(maloaihnh,tenloaihnh) values("{maloaihnh}","{tenloaihnh}");\n'
            file.write(querry)
```

```python
def file_phonggd(dataframe,output_file): # import phonggd
    with open(output_file,"w",encoding="utf-8") as file:
        file.write("use truonghoc;\n")
        for index,row in dataframe.iterrows():
            [tenphng,maphng] = list(row)
            querry = f'INSERT INTO phonggd(maphg,tenphg) values("{maphng}","{tenphng}");\n'
            file.write(querry)
```

```python
def file_truong(dataframe, output_file): # import loaihinh
    with open(output_file, "w", encoding="utf-8") as file:
        file.write("use truonghoc;\n")
        for index, row in dataframe.iterrows():
            [matruong, tentruong, _, _, diachi, maloaihinh, maloaitruong, soggdt, maphong] = list(row)
          
            # Loại bỏ dấu ngoặc kép nếu giá trị là Null
            if pd.isna(diachi):
                diachi = 'NULL'
            else:
                diachi = f'"{diachi}"'
                
            if pd.isna(maloaihinh):
                maloaihinh = 'NULL'
            else:
                maloaihinh = f'"{maloaihinh}"'
                
            if pd.isna(maloaitruong):
                maloaitruong = 'NULL'
            else:
                maloaitruong = f'"{maloaitruong}"'
            
            # Xử lý ngoại lệ cho giá trị `maphong`
            if pd.isna(maphong):
                # Nếu giá trị là Null thì bỏ qua luôn
                continue
            else:
                maphong = f'"{maphong}"'

            # Thêm các giá trị vào câu truy vấn và ghi vào file
            querry = f'INSERT INTO truong(matruong, tentruong, sogddt, maphong, diachi, maloaihinh, maloaitruong) VALUES ("{matruong}", "{tentruong}", "{soggdt}", {maphong}, {diachi}, {maloaihinh}, {maloaitruong});\n'
            file.write(querry)
```

# xử lý data

```python
def process_final(A, B):
    """
    Hàm thêm các mẫu của DataFrame B vào DataFrame A, nếu các giá trị của cột a1 
    đã xuất hiện ở DataFrame A thì không cần thêm mẫu đó.
    
    Parameters:
        A (DataFrame): Dataframe đầu vào A
        B (DataFrame): Dataframe đầu vào B

    Returns:
        DataFrame: Kết quả sau khi thêm các mẫu của DataFrame B vào DataFrame A.
    """
    # Lấy danh sách các giá trị của cột a1 trong DataFrame A
    A_a1_values = set(A["Mã trường"])

    # Lấy ra các mẫu của DataFrame B có giá trị a1 đã xuất hiện trong DataFrame A
    common_B = B[B["Mã trường"].isin(A_a1_values)]

    # Merge DataFrame A với DataFrame common_B theo cột a1 để lấy các mẫu có giá trị của cột a1 đã xuất hiện
    merged_A_common_B = pd.merge(A[["Mã trường"]], common_B, on="Mã trường", how='left')

    # Trích xuất các mẫu của DataFrame B mà không có trong DataFrame merged_A_common_B
    new_B = B[~B.set_index(list(B.columns)).index.isin(merged_A_common_B.set_index(list(merged_A_common_B.columns)).index)]

    # Nối DataFrame A và DataFrame B (chỉ bao gồm các mẫu của B không có trong merged_A_common_B)
    result = pd.concat([A, new_B], ignore_index=True)

    return result
```

```python
df_re = pd.DataFrame(columns=["Mã trường","Tên trường","Sở GD&ĐT ","Phòng GD&DT","Địa chỉ","Loại hình","Loại trường"])
df_re = process_final(df_re,df_thpt)
df_re = process_final(df_re,df_thcs)
df_re = process_final(df_re,df_th)
df_re = process_final(df_re,df_gdtx)
df_re = process_final(df_re,df_mn)
```

```python
# Convert thành một dict
def convert(df_A,atributeA,atributeB):
    df_sub = df_A.set_index(atributeA)
    return df_sub[atributeB].to_dict()
```
